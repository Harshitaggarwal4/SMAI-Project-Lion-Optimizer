```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data provided for LionCustom
data = {
    "LearningRate": [1e-7, 1e-4, 5e-6, 5e-5, 5e-5, 5e-5, 5e-5, 5e-5, 5e-5],
    "WeightDecay": [1, 1, 1, 1, 0.1, 0.25, 0.01, 0.05, 0.2],
    "Train": [0.037, 0.002583, 0.383443, 0.912582, 0.907389, 0.907585, 0.943715, 0.908877, 0.931082],
    "Dev": [0.03, 0.002358, 0.421066, 0.582340, 0.643653, 0.491550, 0.687101, 0.484344, 0.622429]
}

# Creating DataFrame
df = pd.DataFrame(data)

# Pivoting the DataFrame for heatmap compatibility
# We'll create two DataFrames, one for Train and one for Dev
train_pivot = df.pivot("LearningRate", "WeightDecay", "Train")
dev_pivot = df.pivot("LearningRate", "WeightDecay", "Dev")

# Plotting the heatmap for Train
sns.heatmap(train_pivot, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Train Accuracy Heatmap")
plt.show()

# Plotting the heatmap for Dev
sns.heatmap(dev_pivot, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Dev Accuracy Heatmap")
plt.show()

```
